```python
# import libraries
import pandas as pd
from os import listdir
from os.path import isfile, join
import os
import re
import datetime
```

```python
os.chdir('..')
```

```python
data_path = f"0_data/match_data"

# data_path = "/Users/benmurphy/Data/rugby_data/match_data"
data_files = [f for f in listdir(data_path) if isfile(join(data_path, f))]

dfs = []
for data_file in data_files:
    temp = pd.read_csv(f"{data_path}/{data_file}")
    comp = data_file.split('_')[0]

    if comp == 'urc':

        # temp['competition'] = comp
        dfs.append(temp)

df = pd.concat(dfs)

df.head()
```

```python
df.dtypes
```

```python
# date and time
df['match_date_dt'] = pd.to_datetime(df['match_date'])
df['match_date_time_dt'] = pd.to_datetime(df['match_date'] + ' ' + df['match_time_french'])

df = df.sort_values(by=['match_date_dt', 'match_time_french']).reset_index(drop=True)

df
```

```python
df[['match_date_dt', 'home_team', 
    'away_team', 'match_time_french', 
    'match_result']]
```

```python
# df['round_stage'] = df['meta'].apply(lambda x: x.split('Round : ')[1])
# df['round_stage'] = df['round_stage'].apply(lambda x: x.split('  All times shown are french')[0])

# rename teams mispelled
df['home_team'].replace({'Edimbourg': 'Edinburgh',
                         'TrÃ©vise': 'Benetton'}, inplace=True)

df['away_team'].replace({'Edimbourg': 'Edinburgh',
                         'TrÃ©vise': 'Benetton'}, inplace=True)

# make all teams upper case
df['home_team'] = df['home_team'].str.upper()
df['away_team'] = df['away_team'].str.upper()

df[['home_team', 'away_team']].head(5)
```

```python
datetime.datetime.today()

df.loc[df['match_date_dt'] >= datetime.datetime.today()]
```

```python
df['home_score'] = df['match_result'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)
df['away_score'] = df['match_result'].apply(lambda x: re.findall(r'[\d]+', x)[1]).astype(int)

df.loc[df['match_date_dt'] >= datetime.datetime.today(), 'home_score'] = 0
df.loc[df['match_date_dt'] >= datetime.datetime.today(), 'away_score'] = 0

df[['match_result', 'home_score', 'away_score']].tail()
```

```python
# pack weights
df['home_pack_weight_average'].fillna('0', inplace=True)
df['away_pack_weight_average'].fillna('0', inplace=True)

df['home_pack_weight_kg'] = df['home_pack_weight_average'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)
df['away_pack_weight_kg'] = df['away_pack_weight_average'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)

df[['home_pack_weight_average', 'home_pack_weight_kg', 'away_pack_weight_average', 'away_pack_weight_kg']].head()
```

```python
# pack weights
df['home_forwards_average_age'].fillna('0', inplace=True)
df['home_backs_average_age'].fillna('0', inplace=True)
df['home_tallest_player'].fillna('0', inplace=True)

df['away_forwards_average_age'].fillna('0', inplace=True)
df['away_backs_average_age'].fillna('0', inplace=True)
df['away_tallest_player'].fillna('0', inplace=True)

df['home_forwards_average_age'].replace('nan ans', '0', inplace=True)
df['home_backs_average_age'].replace('nan ans', '0', inplace=True)
df['home_tallest_player'].replace('cm', '0', inplace=True)

df['away_forwards_average_age'].replace('nan ans', '0', inplace=True)
df['away_backs_average_age'].replace('nan ans', '0', inplace=True)
df['away_tallest_player'].replace('cm', '0', inplace=True)

df['home_forwards_average_age'] = df['home_forwards_average_age'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)
df['home_backs_average_age'] = df['home_backs_average_age'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)
df['home_tallest_player'] = df['home_tallest_player'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)

df['away_forwards_average_age'] = df['away_forwards_average_age'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)
df['away_backs_average_age'] = df['away_backs_average_age'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)
df['away_tallest_player'] = df['away_tallest_player'].apply(lambda x: re.findall(r'[\d]+', x)[0]).astype(int)


df[['home_forwards_average_age', 'home_backs_average_age', 'home_tallest_player',
   'away_forwards_average_age', 'away_backs_average_age', 'away_tallest_player']].head()
```

```python
# create a season column
mask_24_25 = (df['match_date_dt'] > '2024-07-31') & (df['match_date_dt'] <= '2025-07-11')
mask_23_24 = (df['match_date_dt'] > '2023-07-31') & (df['match_date_dt'] <= '2024-07-11')
mask_22_23 = (df['match_date_dt'] > '2022-07-31') & (df['match_date_dt'] <= '2023-07-11')
mask_21_22 = (df['match_date_dt'] > '2021-07-31') & (df['match_date_dt'] <= '2022-07-11')

df.loc[mask_24_25, "season"] = "24/25"
df.loc[mask_23_24, "season"] = "23/24"
df.loc[mask_22_23, "season"] = "22/23"
df.loc[mask_21_22, "season"] = "21/22"

df[['match_date_dt', 'season']]
```
